1. Types of Data
Data is classified into two main types:

1.1 Qualitative (Categorical) Data
Represents non-numeric characteristics or labels.

Nominal Scale: Categories without a specific order (e.g., eye color, gender, nationality).
Ordinal Scale: Categories with a meaningful order but without a fixed interval (e.g., satisfaction levels: poor, fair, good, excellent).
1.2 Quantitative (Numerical) Data
Represents numeric values with meaningful mathematical operations.

Interval Scale: Numeric values with equal intervals but no true zero (e.g., temperature in Celsius or Fahrenheit).
Ratio Scale: Numeric values with a true zero, allowing meaningful ratios (e.g., height, weight, age, income).

5. Role of Random Sampling
Random sampling ensures that every individual in a population has an equal chance of being selected, making inferences unbiased and representative.

Importance:

Reduces selection bias
Allows generalization to the entire population
Enables statistical tests to be valid
Example:
Polling a random group of voters to predict election outcomes.

6. Skewness and Its Types
Skewness measures data asymmetry.

Right (Positive) Skew: Long right tail, Mean > Median > Mode.
Left (Negative) Skew: Long left tail, Mean < Median < Mode.
Zero Skew: Symmetric distribution, Mean ≈ Median ≈ Mode.
Effect: Skewness influences which measure of central tendency to use. For skewed data, the median is often a better measure of central tendency than the mean.

9. Normal Distribution & Empirical Rule
A normal distribution is bell-shaped and symmetric.

Empirical Rule (68-95-99.7):

68% of values fall within 1 standard deviation.
95% within 2 standard deviations.
99.7% within 3 standard deviations.
Used to determine probabilities and detect anomalies.

11. Random Variables: Discrete vs. Continuous
A random variable (RV) assigns numerical values to outcomes.

Discrete RV: Takes countable values (e.g., number of defective products in a batch).
Continuous RV: Takes infinite values within a range (e.g., weight of an apple).
